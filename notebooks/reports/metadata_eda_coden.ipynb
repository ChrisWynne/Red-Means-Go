```python
import pandas as pd
import numpy as np
import json
```

```python
ROOT_DIR = "../../"
```

```python
with open(ROOT_DIR + "config/config-scraping.json") as f:
    cfg = json.load(f)
```

```python
df = pd.read_csv(ROOT_DIR + cfg["test-metadata-csv-read-path"].format(cfg["selected-game"], cfg["selected-game"]))
```

```python
df
```

```python
basic_analy_cols = ["commentCount", "dislikeCount", "favoriteCount",
                    "likeCount", "viewCount", "z_comments", 
                    "z_dislikes", "z_likes", "z_views"]
```

```python
for key in basic_analy_cols:
    df.plot(x="position", y=key, title="position vs " + key, figsize=(15,5), color="#b82116")
```

```python
z_stats = ["comment", "dislike", "like", "view"]
```

```python
for stat in z_stats:
    df.plot(y="{}Count".format(stat),
            x="z_{}s".format(stat),
            title=stat.capitalize() + "s vs " + stat.capitalize() + "s z-score",
            figsize=(10, 10),
            kind="scatter",
            color="#b82116")
```
